## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

print(df_his)
print(df_geo)
````

````python
# 1) Seu código aqui

# Definindo a quantidade de questões de cada matéria:
qtd_questoes = {'matematica' : 24, 'portugues' : 18, 'geografia' : 8, 'ingles': 8, 'historia': 8, 'fisica' : 12, 'quimica': 12,}

# Calculando a porcentagem média de acertos:
porcentagem_media_mat = df_mat / qtd_questoes['matematica'] * 100

porcentagem_media_por = df_por / qtd_questoes['portugues'] * 100

porcentagem_media_geo = df_geo / qtd_questoes['geografia'] * 100

porcentagem_media_ing = df_ing / qtd_questoes['ingles'] * 100

porcentagem_media_his = df_his / qtd_questoes['historia'] * 100

porcentagem_media_fis = df_fis / qtd_questoes['fisica'] * 100

porcentagem_media_qui = df_qui / qtd_questoes['quimica'] * 100

````

````python
# 2) Seu código aqui

# Definindo a quantidade total de acertos de cada aluno:
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
total_acertos
````

````python
# 3) Seu código aqui

# Definindo a quantidade total de questões:

lista = []
for materia, valor in qtd_questoes.items():
    lista.append(valor)
    
qtd_questoes_total = sum(lista)

# Calculando a porcentagem média de acertos com relação ao total de questões:
porcentagem_media_acertos_total = total_acertos / qtd_questoes_total * 100

````

````python
# 4) Seu código aqui

aprovado = porcentagem_media_acertos_total >= 45
qtd_aprovados = aprovado.sum()
qtd_aprovados
````

````python
# 2) 

# Definir uma variável para minha coluna
paciente_id = arquivo_base['paciente_id']

pacientes_diferentes = paciente_id.nunique()
print(f"Quantidade de pacientes diferentes: {pacientes_diferentes}")
````

````python
# 3) 

# Criando um boleeano que vai identificar a quantidade de pacientes que tomaram a primeira dose:

qtd_primeiradose_distribuida = arquivo_base['vacina_descricao_dose'] == '1ª Dose'

# Criando um DataFrame
df_01 = pd.DataFrame(qtd_primeiradose_distribuida)
df_01


# Contando a quantidade de pacientes que tomaram a primeira dose (mensurado pela quantidade de "True"):
contagem_primeira_dose = df_01['vacina_descricao_dose'].value_counts()
print(contagem_primeira_dose)
````

````python
# 4) 


# Criando um boleeano que vai identificar o paciente que tem menos de 18 anos como True:

menores_18anos = arquivo_base['paciente_idade'] < 18

# Criando um DataFrame
df_02 = pd.DataFrame(menores_18anos)
df_02


# Contando a quantidade de pacientes com menos de 18 anos que foram vacinados (mensurado pela quantidade de "True"):
contagem_menor18anos = df_02['paciente_idade'].value_counts()
print(contagem_menor18anos)

````
